```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

data = np.loadtxt('Yb_fit.txt')

i = 85
x_values = data[(360)*(i-1)+101:(360)*(i-1)+140,0]
y_values = data[(360)*(i-1)+101:(360)*(i-1)+140,1]

i = 85
x_values1 = data[(360)*(i-1)+81:(360)*(i-1)+160,0]
y_values1 = data[(360)*(i-1)+81:(360)*(i-1)+160,1]

N = np.argmin(y_values)
x_shift = []
for i in range(0,len(x_values1)):
    x_new = x_values1[i]-x_values[N]
    x_shift.append(x_new)

x = x_shift
y = y_values1

# Define a 4th order polynomial function with a fixed constant term
def poly_4th_order(x, a, b, c, d):
    return a * x**4 + b * x**3 + c * x**2 + d * x - 0.0163177

# Initial guess for the coefficients
initial_guess = [1, 1, 1, 1]

# Fit the polynomial function to the data
popt, pcov = curve_fit(poly_4th_order, x, y, p0=initial_guess)

# Generate points for the fitted polynomial curve
x_fit = np.linspace(min(x), max(x), 100)
y_fit = poly_4th_order(x_fit, *popt)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data')
plt.plot(x_fit, y_fit, 'r-', label='4th Order Polynomial Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.title('4th Order Polynomial Fit with Constant Term -0.017')
plt.legend()
plt.grid(True)
plt.show()

print("Parameters of the polynomial fit (a, b, c, d):")
print(popt)
```
